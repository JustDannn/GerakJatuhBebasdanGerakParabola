Suatu Meriam ditembakan dengan sudut kemiringan 60 derajat, kecepatan awal dari peluru yang ditembakan adalah 10m/s.Berdasarkan perhitungan 
-	Tentukan waktu yang diperlukan oleh peluru untuk mencapai ketinggian 2/3 dari 
      ketinggian maksimum
-	Tentukan waktu yang diperlukan oleh peluru untuk mencapai 2/3 jarak 
                          maksimum


```python
import numpy as np
import matplotlib.pyplot as plt

# Data yang diberikan
v0 = 10  # kecepatan awal (m/s)
theta = 60  # sudut kemiringan (derajat)
g = 9.8  # percepatan gravitasi (m/s^2)

# Konversi sudut ke radian
theta_rad = np.radians(theta)

# Komponen kecepatan awal
v0x = v0 * np.cos(theta_rad)  # kecepatan horizontal
v0y = v0 * np.sin(theta_rad)  # kecepatan vertikal
```

```python
print(f"Data Awal:")
print(f"  Kecepatan awal (v0) = {v0} m/s")
print(f"  Sudut elevasi (θ) = {theta}°")
print(f"  Percepatan gravitasi (g) = {g} m/s²")
```

```python
print(f"Komponen Kecepatan:")
print(f"  v0x = v0 cos(θ) = {v0x:.4f} m/s")
print(f"  v0y = v0 sin(θ) = {v0y:.4f} m/s")
```

```python
# 1. MENGHITUNG KETINGGIAN MAKSIMUM DAN JARAK MAKSIMUM

# Ketinggian maksimum: H_max = (v_{0}^y)^2 / (2g)
H_max = (v0y ** 2) / (2 * g)

# Waktu untuk mencapai ketinggian maksimum: t_max_height = v_{0}^y / g
t_max_height = v0y / g

# Jarak maksimum (range): R_max = (v_0^2 sin(2θ)) / g
R_max = (v0 ** 2 * np.sin(2 * theta_rad)) / g

# Waktu total di udara: t_total = 2 * v_{0}^y / g
t_total = 2 * v0y / g
```

```python

print(f"  Ketinggian maksimum (H_max) = {H_max:.4f} m")
print(f"  Waktu mencapai H_max = {t_max_height:.4f} s")
print(f"  Jarak maksimum (R_max) = {R_max:.4f} m")
print(f"  Waktu total di udara = {t_total:.4f} s")
```

```python
# 2. WAKTU UNTUK MENCAPAI 2/3 KETINGGIAN MAKSIMUM

# Target ketinggian
h_target = (2/3) * H_max

print("\n2. WAKTU MENCAPAI 2/3 KETINGGIAN MAKSIMUM:")
print(f"  Target ketinggian = (2/3) × {H_max:.4f} = {h_target:.4f} m")

a = g / 2
b = -v0y
c = h_target

# Menggunakan rumus kuadrat: t = (-b ± \sqrt{(b^2-4ac)}) / 2a
discriminant = b**2 - 4*a*c
t1 = (-b + np.sqrt(discriminant)) / (2*a)
t2 = (-b - np.sqrt(discriminant)) / (2*a)
```

```python

print(f"Menyelesaikan persamaan: {a:.1f}t^2 + ({b:.4f})t + {c:.4f} = 0")
```

```python

print(f"Solusi kuadrat:")
print(f"    t_1 = {t1:.4f} s (saat naik)")
print(f"    t_2 = {t2:.4f} s (saat turun)")
```

```python
print(f"Peluru mencapai ketinggian {h_target:.4f} m pada:")
print(f"    - t = {t1:.4f} s (fase naik)")
print(f"    - t = {t2:.4f} s (fase turun)")
```

```python
# 3. WAKTU UNTUK MENCAPAI 2/3 JARAK MAKSIMUM

# Target jarak horizontal
x_target = (2/3) * R_max

print("\n3. WAKTU MENCAPAI 2/3 JARAK MAKSIMUM:")
print(f"  Target jarak horizontal = (2/3) × {R_max:.4f} = {x_target:.4f} m")

# Persamaan gerak horizontal: x(t) = v0x·t
# Maka: t = x / v0x
t_2_3_range = x_target / v0x

print(f"\n  Dari persamaan: x = v0x t")
print(f"  Maka: t = x / v0x = {x_target:.4f} / {v0x:.4f}")
print(f"\n   Peluru mencapai jarak {x_target:.4f} m pada:")
print(f"    t = {t_2_3_range:.4f} s")

# Hitung ketinggian pada saat mencapai 2/3 jarak maksimum
y_at_2_3_range = v0y * t_2_3_range - 0.5 * g * t_2_3_range**2
print(f"\n  Pada saat ini, ketinggian peluru = {y_at_2_3_range:.4f} m")
```

```python
print("Pertanyaan 1: Waktu untuk mencapai 2/3 ketinggian maksimum")
print(f"  Ketinggian target: {h_target:.4f} m")
print(f"  Jawaban:")
print(f" Saat naik  : t = {t1:.4f} s")
print(f" Saat turun : t = {t2:.4f} s")
```

```python

print("Pertanyaan 2: Waktu untuk mencapai 2/3 jarak maksimum")
print(f"  Jarak target: {x_target:.4f} m")
print(f"  Jawaban:")
print(f" t = {t_2_3_range:.4f} s")
print(f" Ketinggian saat itu: {y_at_2_3_range:.4f} m")
```

```python
# 5. VISUALISASI TRAJEKTORI

# Membuat array waktu untuk simulasi
t_array = np.linspace(0, t_total, 500)

# Menghitung posisi x dan y untuk setiap waktu
x_array = v0x * t_array
y_array = v0y * t_array - 0.5 * g * t_array**2

# Membuat plot
plt.figure(figsize=(14, 8))

# Plot trajektori lengkap
plt.plot(x_array, y_array, 'b-', linewidth=2, label='Trajektori peluru')

# Menandai titik-titik penting
# Titik awal
plt.plot(0, 0, 'go', markersize=12, label='Titik tembak', zorder=5)

# Ketinggian maksimum
x_at_max_h = v0x * t_max_height
plt.plot(x_at_max_h, H_max, 'ro', markersize=12, label=f'H_max = {H_max:.2f} m', zorder=5)

# 2/3 ketinggian maksimum (saat naik dan turun)
x_at_t1 = v0x * t1
x_at_t2 = v0x * t2
plt.plot(x_at_t1, h_target, 'mo', markersize=10, label=f'2/3 H_max (naik, t={t1:.3f}s)', zorder=5)
plt.plot(x_at_t2, h_target, 'co', markersize=10, label=f'2/3 H_max (turun, t={t2:.3f}s)', zorder=5)

# 2/3 jarak maksimum
plt.plot(x_target, y_at_2_3_range, 'ys', markersize=12, 
         label=f'2/3 R_max (t={t_2_3_range:.3f}s, y={y_at_2_3_range:.2f}m)', zorder=5)

# Titik jatuh
plt.plot(R_max, 0, 'ko', markersize=10, label=f'Titik jatuh (R_max = {R_max:.2f} m)', zorder=5)

# Garis bantu horizontal untuk 2/3 ketinggian maksimum
plt.axhline(y=h_target, color='purple', linestyle='--', alpha=0.3, linewidth=1)
plt.text(R_max*0.05, h_target + 0.3, f'y = 2/3 H_max = {h_target:.2f} m', 
         fontsize=10, color='purple')

# Garis bantu vertikal untuk 2/3 jarak maksimum
plt.axvline(x=x_target, color='orange', linestyle='--', alpha=0.3, linewidth=1)
plt.text(x_target + 0.3, H_max*0.9, f'x = 2/3 R_max\n= {x_target:.2f} m', 
         fontsize=10, color='orange')

# Pengaturan plot
plt.xlabel('Jarak Horizontal (m)', fontsize=12, fontweight='bold')
plt.ylabel('Ketinggian (m)', fontsize=12, fontweight='bold')
plt.title('Trajektori Gerak Parabola Peluru Meriam\n' + 
          f'v₀ = {v0} m/s, θ = {theta}^o, g = {g} m/s²', 
          fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, linestyle=':')
plt.legend(loc='upper right', fontsize=9)
plt.xlim(-0.5, R_max + 1)
plt.ylim(-0.5, H_max + 1)
plt.tight_layout()
plt.show()
```

## Penjelasan Rumus dan Metode

### Rumus-rumus yang Digunakan:

**1. Komponen Kecepatan:**
- $v_{0x} = v_0 \cos(\theta)$ (kecepatan horizontal)
- $v_{0y} = v_0 \sin(\theta)$ (kecepatan vertikal)

**2. Ketinggian Maksimum:**
$$H_{max} = \frac{v_{0y}^2}{2g}$$

**3. Jarak Maksimum (Range):**
$$R_{max} = \frac{v_0^2 \sin(2\theta)}{g}$$

**4. Persamaan Gerak:**
- Horizontal: $x(t) = v_{0x} \cdot t$
- Vertikal: $y(t) = v_{0y} \cdot t - \frac{1}{2}g t^2$

**5. Untuk mencari waktu pada ketinggian tertentu:**

Dari persamaan vertikal: $h = v_{0y} \cdot t - \frac{1}{2}g t^2$

Diatur ulang: $\frac{1}{2}g t^2 - v_{0y} t + h = 0$

Menggunakan rumus kuadrat: $t = \frac{v_{0y} \pm \sqrt{v_{0y}^2 - 2gh}}{g}$
