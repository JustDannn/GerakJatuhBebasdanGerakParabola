# SOAL NO 1 
Tinjaulah suatu kejadian gerak jatuh bebas. Suatu kelapa terjatuh dari pohonnya yang memiliki ketinggian 10m. Tentukan waktu maksimum andi dengan tinngi 160 cm untuk bereaksi supaya tidak tertimpa buah kelapa yang jatuh

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
tinggi_pohon = 10  # meter
tinggi_andi = 1.6 # meter
gravitasi = 9.8  # m/s^2 (Nilai diubah sesuai permintaan)

# Hitung jarak yang ditempuh kelapa hingga mencapai kepala Andi
jarak_jatuh = tinggi_pohon - tinggi_andi

# Hitung waktu reaksi (waktu untuk mencapai tinggi Andi)
waktu_reaksi = math.sqrt((2 * jarak_jatuh) / gravitasi)

# Hitung waktu total untuk jatuh ke tanah
waktu_total_jatuh = math.sqrt((2 * tinggi_pohon) / gravitasi)

# Menampilkan hasil perhitungan teks
print(f"Tinggi pohon: {tinggi_pohon} m")
print(f"Tinggi Andi: {tinggi_andi} m")
print(f"Jarak jatuh kelapa hingga mengenai Andi: {jarak_jatuh:.2f} m")
print("-" * 30)
print(f"Waktu reaksi maksimum yang dimiliki Andi adalah: {waktu_reaksi:.2f} detik")
```

```python
# Membuat rentang waktu dari 0 hingga waktu total jatuh
t = np.linspace(0, waktu_total_jatuh, 100)

# Menghitung ketinggian pada setiap waktu t
h = tinggi_pohon - 0.5 * gravitasi * t**2

# --- 3. Membuat Plot ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(10, 6))

# Plot utama
ax.plot(t, h, label='Lintasan Kelapa', color='brown', linewidth=2.5)

# Menandai titik kritis
ax.plot(waktu_reaksi, tinggi_andi, 'o', markersize=10, color='red', label=f'Waktu Reaksi Maksimal ({waktu_reaksi:.2f} s)')

# Garis bantu
ax.axhline(y=tinggi_andi, color='red', linestyle='--', linewidth=1, label=f'Tinggi Andi ({tinggi_andi} m)')
ax.axvline(x=waktu_reaksi, color='red', linestyle='--', linewidth=1)

# --- 4. Kustomisasi Plot ---
ax.set_title('Grafik Gerak Jatuh Bebas Kelapa (g = 9,8 m/s)', fontsize=16, fontweight='bold')
ax.set_xlabel('Waktu (detik)', fontsize=12)
ax.set_ylabel('Ketinggian (meter)', fontsize=12)
ax.legend(fontsize=10)
ax.set_xlim(0, waktu_total_jatuh * 1.05)
ax.set_ylim(0, tinggi_pohon * 1.05)
plt.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Inisialisasi Konstanta dan Parameter ---
tinggi_awal = 10  # meter
gravitasi = 10    # m/s^2
dt = 0.1          # Langkah waktu (time step) dalam detik untuk simulasi numerik

# --- 2. Solusi Analitik ---
# Waktu total untuk jatuh (solusi eksak)
waktu_total_analitik = np.sqrt((2 * tinggi_awal) / gravitasi)
# Membuat 100 titik waktu untuk kurva yang mulus
waktu_analitik = np.linspace(0, waktu_total_analitik, 100)
# Menghitung ketinggian analitik di setiap titik waktu
tinggi_analitik = tinggi_awal - 0.5 * gravitasi * waktu_analitik**2

# --- 3. Solusi Numerik (Metode Euler) ---
# List untuk menyimpan hasil simulasi
waktu_numerik = [0]
tinggi_numerik = [tinggi_awal]
kecepatan_numerik = 0
t_numerik = 0
h_numerik_step = tinggi_awal

# Loop simulasi hingga objek menyentuh tanah (ketinggian > 0)
while h_numerik_step > 0:
    # Update posisi menggunakan kecepatan *sebelumnya*
    h_numerik_step = h_numerik_step - kecepatan_numerik * dt
    # Update kecepatan untuk langkah berikutnya
    kecepatan_numerik = kecepatan_numerik + gravitasi * dt
    # Update waktu
    t_numerik += dt
    
    # Simpan hasil
    waktu_numerik.append(t_numerik)
    tinggi_numerik.append(h_numerik_step)

# --- 4. Plot Perbandingan ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(10, 6))

# Plot solusi analitik sebagai garis mulus
ax.plot(waktu_analitik, tinggi_analitik, 'b-', label='Solusi Analitik (Eksak)', linewidth=2.5)
# Plot solusi numerik sebagai titik-titik
ax.plot(waktu_numerik, tinggi_numerik, 'ro--', label=f'Solusi Numerik (dt={dt} s)', markersize=6, linestyle='--')

ax.set_title('Perbandingan Solusi Analitik vs. Numerik', fontsize=16, fontweight='bold')
ax.set_xlabel('Waktu (detik)', fontsize=12)
ax.set_ylabel('Ketinggian (meter)', fontsize=12)
ax.legend()
ax.set_xlim(0, waktu_total_analitik * 1.05)
ax.set_ylim(0, tinggi_awal * 1.05)

plt.tight_layout()
plt.savefig('perbandingan_analitik_numerik.png')
# plt.show()
```

### Langkah-langkah Perhitungan



1.  **Menentukan Data yang Diketahui**
    * Ketinggian total pohon ($h_{total}$): 10 m
    * Tinggi Andi ($h_{Andi}$): 160 cm = 1,6 m
    * Percepatan gravitasi ($g$): 9,8 m/s

2.  **Menghitung Jarak Jatuh Kelapa**
    Kelapa tidak perlu sampai ke tanah untuk mengenai Andi. Kita hanya perlu menghitung waktu tempuh hingga kelapa mencapai puncak kepala Andi.
    * Jarak jatuh ($h_{jatuh}$) = Ketinggian pohon - Tinggi Andi
    * $h_{jatuh} = 10 \ m - 1,6 \ m = 8,4 \ m$

3.  **Menghitung Waktu Reaksi (Waktu Jatuh)**
    Kita menggunakan rumus gerak jatuh bebas untuk mencari waktu ($t$).
    * Rumus: $t = \sqrt{\frac{2h}{g}}$
    * Masukkan nilai yang diketahui:
        $t = \sqrt{\frac{2 \cdot 8,4}{9,8}}$
    * Lakukan perhitungan:
        $t = \sqrt{\frac{16,8}{9,8}}$
        $t = \sqrt{1,714}$
        $t \approx 1,309 \ s$

Jadi, Andi memiliki waktu sekitar **1,31 detik** untuk menghindar setelah kelapa mulai jatuh agar tidak tertimpa.
